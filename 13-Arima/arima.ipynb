```python
# fit an ARIMA model and plot residual errors
from datetime import datetime
from pandas import read_csv
from pandas import DataFrame
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot
# load dataset
def parser(x):
	return datetime.strptime('190'+x, '%Y-%m')
series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True, date_parser=parser)
series.index = series.index.to_period('M')
# fit model
model = ARIMA(series, order=(5,1,0))
model_fit = model.fit()
# summary of fit model
print(model_fit.summary())
# line plot of residuals
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()
# summary stats of residuals
print(residuals.describe())
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def forward_propagation(inputs, weights, bias):
    # Calculate the weighted sum of inputs plus bias
    weighted_sum = np.dot(inputs, weights) + bias
    # Apply the sigmoid activation function
    output = sigmoid(weighted_sum)
    return output

# Example usage:
# Define the inputs, weights, and bias
inputs = np.array([1, 1])
weights = np.array([1, 1])
bias = 1

# Perform forward propagation
output = forward_propagation(inputs, weights, bias)
print(f"The output of the neural network is: {output}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNeuralNetwork, self).__init__()
        # Define the layers of the network
        self.fc1 = nn.Linear(input_size, hidden_size)  # First fully connected layer
        self.relu = nn.ReLU()                         # ReLU activation function
        self.fc2 = nn.Linear(hidden_size, output_size) # Second fully connected layer

    def forward(self, x):
        # Define the forward pass (how data flows through the network)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Example usage:
input_dim = 784  # e.g., for flattened 28x28 images
hidden_dim = 128
output_dim = 10  # e.g., for 10 classes in classification

# Create an instance of the neural network
model = SimpleNeuralNetwork(input_dim, hidden_dim, output_dim)

# Print the model architecture
print(model)

# Create a dummy input tensor
dummy_input = torch.randn(1, input_dim) # Batch size of 1

# Perform a forward pass
output = model(dummy_input)
print("Output shape:", output.shape)
```
