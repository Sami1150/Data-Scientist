```python
from matplotlib import pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 11]

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.title('Simple Line Plot')
plt.xlabel('X-values')
plt.ylabel('Y-values')    
plt.grid(True)
plt.legend(['Data Points'])

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
x_values = [6, 7, 8, 9, 10]
y_values = [22, 23, 25, 27, 22]
plt.xticks(x_values)
plt.yticks(y_values)
plt.xlim(0, 11)
plt.ylim(0, 28)
plt.axhline(0, color='black', linewidth=11, linestyle='--')
plt.axvline(0, color='black', linewidth=11, linestyle='--')

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

plt.savefig('simple_line_plot.png')
plt.show()
```

```python
from matplotlib import pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 11]

plt.scatter(x_values, y_values, marker='*', linestyle='-', color='b')
plt.title('Simple Line Plot')
plt.xlabel('X-values')
plt.ylabel('Y-values')    
plt.grid(True)
plt.legend(['Data Points'])
plt.savefig('simple_line_plot.png')
plt.show()
```
